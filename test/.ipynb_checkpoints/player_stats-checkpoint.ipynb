```python
import pandas as pd
from matplotlib import pyplot as plt

%matplotlib inline
```

```python
players = {}


for i in list(range(9)):
    players[f'player{i}'] = pd.read_csv(f'player{i}.csv')

def plot_frame(df, i):
    plt.subplot(3, 3, i+1)
    plt.plot(df.level, df.health, label='Health')
    plt.plot(df.level, df.mana, label='Mana')
    plt.plot(df.level, df.physicalAttack, label='Physical Attack')
    plt.plot(df.level, df.physicalDefense, label='Physical Defense')
    plt.plot(df.level, df.magicAttack, label='Magic Attack')
    plt.plot(df.level, df.magicDefense, label='Magic Defense')
    plt.plot(df.level, df.speed, label='Speed')
    plt.plot(df.level, df.luck, label='Luck')
    plt.title(f'Player {i+1}')

def plot_all_frames():
    plt.figure(1, figsize=(20, 20), edgecolor='blue')
    for i in list(range(9)):
        plot_frame(players[f'player{i}'], i)
    plt.subplots_adjust(hspace=1, wspace=1)
    plt.legend()
    plt.show()
```

```python
plot_all_frames()
```

```python
from matplotlib import pyplot as plt

x = range(7)
straight_line = [0, 1, 2, 3, 4, 5, 6]
parabola = [0, 1, 4, 9, 16, 25, 36]
cubic = [0, 1, 8, 27, 64, 125, 216]

# Subplot 1
plt.subplot(2, 1, 1)
plt.plot(x, straight_line)

# Subplot 2
plt.subplot(2, 2, 3)
plt.plot(x, parabola)

# Subplot 3
plt.subplot(2, 2, 4)
plt.plot(x, cubic)

plt.subplots_adjust(wspace=0.5, bottom=0.2, hspace=0.5)

plt.show()
```
