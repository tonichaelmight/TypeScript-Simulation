```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
plt.rc('font', size=12)
plt.rc('lines', lw=3)
plt.rc('xtick', labelsize=14)
plt.rc('ytick', labelsize=14)
plt.rc('axes', titlesize=24, labelsize=18)
```

```python
players = {}

for i in list(range(9)):
  players[f'player{i}'] = pd.read_csv(f'player{i}.csv')

def plot_stats(df, i):
  plt.subplot(3, 3, i+1)
  plt.plot(df.level, df.health, label=f'Health {df.health.max()}')
  plt.plot(df.level, df.mana, label=f'Mana {df.mana.max()}')
  plt.plot(df.level, df.physicalAttack, label=f'P Attack {df.physicalAttack.max()}')
  plt.plot(df.level, df.physicalDefense, label=f'P Defense {df.physicalDefense.max()}')
  plt.plot(df.level, df.magicAttack, label=f'M Attack {df.magicAttack.max()}')
  plt.plot(df.level, df.magicDefense, label=f'M Defense {df.magicDefense.max()}')
  plt.plot(df.level, df.speed, label=f'Speed {df.speed.max()}')
  plt.plot(df.level, df.luck, label=f'Luck {df.luck.max()}')
  plt.xlabel('Level')
  plt.ylabel('Stat Value')
  plt.title(f'Player {i+1} Stats')
  plt.legend()

def plot_all_stats():
  plt.subplots(3, 3, num=1, figsize=(45, 45), edgecolor='red', facecolor='skyblue')
  for i in list(range(9)):
    plot_stats(players[f'player{i}'], i)
  plt.subplots_adjust(hspace=0.25, wspace=0.25)
  plt.show()
```

```python
plot_all_stats()
```

```python
players = {}

for i in list(range(9)):
  players[f'player{i}'] = pd.read_csv(f'player{i}.csv')
  
def plot_matrix(df, i):
  plt.subplot(3, 3, i+1)
  plt.plot(df.level, df.health_matrix, label=f'Health {df.health_matrix.max()}')
  plt.plot(df.level, df.mana_matrix, label=f'Mana {df.mana_matrix.max()}')
  plt.plot(df.level, df.physicalAttack_matrix, label=f'P Attack {df.physicalAttack_matrix.max()}')
  plt.plot(df.level, df.physicalDefense_matrix, label=f'P Defense {df.physicalDefense_matrix.max()}')
  plt.plot(df.level, df.magicAttack_matrix, label=f'M Attack {df.magicAttack_matrix.max()}')
  plt.plot(df.level, df.magicDefense_matrix, label=f'M Defense {df.magicDefense_matrix.max()}')
  plt.plot(df.level, df.speed_matrix, label=f'Speed {df.speed_matrix.max()}')
  plt.plot(df.level, df.luck_matrix, label=f'Luck {df.luck_matrix.max()}')
  plt.xlabel('Level')
  plt.ylabel('Matrix Value')
  plt.title(f'Player {i+1} Leveling Matrix')
  plt.legend()

def plot_all_matrices():
  plt.close('all')
  plt.subplots(3, 3, num=2, figsize=(45, 45), edgecolor='blue', facecolor=(0.7, 0.45, 0.8))
  for i in list(range(9)):
    plot_matrix(players[f'player{i}'], i)
  plt.subplots_adjust(hspace=0.25, wspace=0.25)
  plt.show()
```

```python

plot_all_matrices()
```

```python
players = {}

for i in list(range(9)):
  players[f'player{i}'] = pd.read_csv(f'player{i}.csv')

stats = ['health', 'mana', 'physicalAttack', 'physicalDefense', 'magicAttack', 'magicDefense', 'speed', 'luck']

def plot_stat(stat, i):
  plt.subplot(3, 3, i+1)

  for i in list(range(9)):
    plt.plot(players[f'player{i}'].level, players[f'player{i}'][stat], label=f'player {i+1}')  

  plt.xlabel('Level')
  plt.ylabel('Stat Value')
  plt.title(stat.title())
  plt.legend()

def plot_all_stats():
  plt.close('all')
  plt.subplots(3, 3, num=3, figsize=(45, 45), edgecolor='blue', facecolor='skyblue')
  for i in list(range(len(stats))):
    plot_stat(stats[i], i)
  plt.subplots_adjust(hspace=0.25, wspace=0.25)
  plt.show()
```

```python
plot_all_stats()
```
