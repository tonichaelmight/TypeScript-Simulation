```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
players = {}

for i in list(range(9)):
  players[f'player{i}'] = pd.read_csv(f'player{i}.csv')

def plot_stats(df, i):
  plt.subplot(3, 3, i+1)
  plt.plot(df.level, df.health, label='Health')
  plt.plot(df.level, df.mana, label='Mana')
  plt.plot(df.level, df.physicalAttack, label='Physical Attack')
  plt.plot(df.level, df.physicalDefense, label='Physical Defense')
  plt.plot(df.level, df.magicAttack, label='Magic Attack')
  plt.plot(df.level, df.magicDefense, label='Magic Defense')
  plt.plot(df.level, df.speed, label='Speed')
  plt.plot(df.level, df.luck, label='Luck')
  plt.title(f'Player {i+1} Stats')

def plot_all_stats():
  plt.subplots(3, 3, num=1, figsize=(20, 20), edgecolor='blue', facecolor='skyblue')
  for i in list(range(9)):
    plot_stats(players[f'player{i}'], i)
  plt.subplots_adjust(hspace=1, wspace=1)
  plt.legend()
  plt.show()
```

```python
plot_all_stats()
```

```python
players = {}

for i in list(range(9)):
  players[f'player{i}'] = pd.read_csv(f'player{i}.csv')
  
def plot_matrices(df, i):
  plt.subplot(3, 3, i+1)
  plt.plot(df.level, df.health_matrix, label='Health')
  plt.plot(df.level, df.mana_matrix, label='Mana')
  plt.plot(df.level, df.physicalAttack_matrix, label='Physical Attack')
  plt.plot(df.level, df.physicalDefense_matrix, label='Physical Defense')
  plt.plot(df.level, df.magicAttack_matrix, label='Magic Attack')
  plt.plot(df.level, df.magicDefense_matrix, label='Magic Defense')
  plt.plot(df.level, df.speed_matrix, label='Speed')
  plt.plot(df.level, df.luck_matrix, label='Luck')
  plt.title(f'Player {i+1} Leveling Matrix')

def plot_all_matrices():
  plt.close('all')
  plt.subplots(3, 3, num=2, figsize=(20, 20), edgecolor='blue', facecolor=(0.5, 0.3, 0.6))
  for i in list(range(9)):
    plot_stats(players[f'player{i}'], i)
  plt.subplots_adjust(hspace=1, wspace=1)
  plt.legend()
  plt.show()
```

```python

plot_all_matrices()
```
